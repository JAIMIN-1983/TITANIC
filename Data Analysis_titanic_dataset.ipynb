# Data Cleaning - Remove null Values

```python
plt.figure(figsize=(12,7))
sns.boxplot(x= 'Pclass', y = 'Age', data = df, palette = 'viridis_r')
```

```python
df.isnull().sum()
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
df['Age']=df[['Age', 'Pclass']].apply(impute_age, axis = 1)
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), yticklabels=False)
```
